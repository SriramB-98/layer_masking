```python
def block_segment(imgs, patch_size=16):
    seg_num = 224//patch_size
    segments = np.arange(seg_num*seg_num).reshape((seg_num, 1, seg_num, 1))
    seg_mask =np.tile(segments, (1, patch_size, 1, patch_size))
    seg_mask = seg_mask.reshape(224, 224)
    return seg_mask
```

```python
quickshift_seg = lambda x : quickshift(x,kernel_size=2,
                                    max_dist=200, ratio=0.2,
                                    random_seed=0)
square_patch_seg = lambda x: block_segment(x, patch_size=16)
slic_seg = lambda x: slic(x,  n_segments=200,)
```

```python
def get_edges(segs):
    return ((np.roll(segs, 1, axis=0) != segs) | (np.roll(segs, 1, axis=1) != segs) )[:,:,None]
```

```python
def turn_off(ax):
    ax.tick_params(
                axis='both',          # changes apply to the x-axis
                which='both',      # both major and minor ticks are affected
                bottom=False,      # ticks along the bottom edge are off
                top=False,         # ticks along the top edge are off
                labelbottom=False) 
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    return
```

```python
from torchvision.models import resnet50
model = resnet50(pretrained=True)
```

```python
def part_conv(model, inp):
    out = model.conv1(inp)
    out = model.bn1(out)
    out = model.relu(out)
    out = model.maxpool(out)
    
    out = model.layer1(out)
#     out = model.layer2(out)
#     out = model.layer3(out)
    if isinstance(out, tuple):
        out = out[0]
    return out
```

```python
from torch.nn.functional import conv2d
def fast_fmm(img, mask, iters=5, eps=1e-5):
    filt = torch.ones(1, 1, 3, 3)
    img = img*mask
    c = img.shape[1]
    filtc = filt.expand(c, -1, -1, -1)
    for i in range(iters):
        with torch.no_grad():
            num = conv2d(img, filtc, groups=c, padding='same')
            denom = conv2d(mask, filt, groups=1, padding='same')
            edge_fills = (1-mask)*(num/(denom+eps))
        img = img + edge_fills.detach()
        mask = (mask + denom).clip(0,1)
    return img
```
